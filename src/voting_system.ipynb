```python
import pandas as pd
```

```python
res1 = pd.read_csv('./texts/cnn_gru_ekphrasis.txt', sep='\t')
res2 = pd.read_csv('./texts/cnn_lstm_simple_preprocessing.txt', sep='\t')['label'].tolist()
res3 = pd.read_csv('./texts/gru_cnn_dropout04_ekphrasis.txt', sep='\t')['label'].tolist()
res4 = pd.read_csv('./texts/gru_trainable_simple_preprocessing.txt', sep='\t')['label'].tolist()
res5 = pd.read_csv('./texts/lstm_cnn_trainable_dropout02_simple_preprocessing.txt', sep='\t')['label'].tolist()
```

```python
res_zipped = list(zip(res1['label'].tolist(), res2, res3, res4, res5))
```

```python
res_zipped
```

```python
vote = list(map(lambda x: max(set(x), key=x.count), res_zipped))
```

```python
vote
```

```python
res1['label'] = vote
```

```python
res1.to_csv('./test.txt', sep='\t', index=False)
```

```python
# for binary classification
angry = pd.read_csv('./tests/binary_angry.txt', sep='\t')
sad = pd.read_csv('./tests/binary_sad.txt', sep='\t')['label'].tolist()
happy = pd.read_csv('./tests/binary_happy.txt', sep='\t')['label'].tolist()
res_zipped = list(zip(angry['label'].tolist(), sad, happy))
vote = list(map(lambda x: min(set(x), key=x.count), res_zipped))
angry['label'] = vote
angry.to_csv('./test.txt', sep='\t', index=False)
```
